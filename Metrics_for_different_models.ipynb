```python
# (Model name, Inference time, mAP 50)
inference_time = [[("Faster R-CNN ResNet50 C4 x1", 201.92, 0.965)], [("EfficientDet", 40.75, 0.915)], [("YOLOv8 nano", 8.3, 0.986), ("YOLOv8 small", 18.9, 0.990)]] # All in miliseconds
```

```python
import plotly.graph_objects as go

# Create an empty figure
fig = go.Figure()

# Iterate through each list of tuples
for models in inference_time:
    # Check the number of models in the current list
    if len(models) == 1:
        # If there is one model, plot a dot
        model_name, inf_time, mAP50 = models[0]
        fig.add_trace(go.Scatter(
            x=[inf_time],
            y=[mAP50],
            mode='markers',
            name=model_name,
            marker=dict(size=8)
        ))
    elif len(models) == 2:
        # If there are two models, plot a line plot with markers
        x = []
        y = []
        names = []
        for model in models:
            model_name, inf_time, mAP50 = model
            x.append(inf_time)
            y.append(mAP50)
            names.append(model_name)

        # Add the line plot with markers to the figure
        fig.add_trace(go.Scatter(
            x=x,
            y=y,
            mode='lines+markers',
            name=f"{names[0]} & {names[1]}",
            marker=dict(size=8)  # Customize marker size as desired
        ))

# Add labels and title to the plot
fig.update_layout(
    xaxis_title="Inference Time (ms)",
    yaxis_title="mAP 50",
    title="Model Performance",
    showlegend=True
)

# Show the plot
fig.show()
```
